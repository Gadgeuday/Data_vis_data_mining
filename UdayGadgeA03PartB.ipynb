We can look at job description and build a word cloud to get an overview of the requirements.

```python
jobs[['Job Description']]
```

First step would be to extract the text from the web link. We can do this using regular expressions by substituting everything in '<' and '>' with ''.

```python
import re
jobs['Job Description'] = jobs['Job Description'].apply(lambda x: re.sub('<.*?>', '', x))
jobs[['Job Description']]
```

We can convert the text to lower case and removing special characters.

```python
jobs['clean_text'] = jobs['Job Description'].apply(lambda x: re.sub('[^A-Za-z]+', ' ', str(x)).lower())
jobs.head()
```

Now let's build the word cloud

```python
#!pip install wordcloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image

# let's take words that are longer than 3 words
text = jobs['clean_text'].apply(lambda x: ' '.join([i for i in x.split(" ") if (len(i) > 3)]))
text = ' '.join([i for i in text.tolist() if len(i) > 3])

# it is important to take stop words out of word cloud
wordcloud = WordCloud(stopwords = STOPWORDS,
                      collocations=True,
                      width = 1000, height = 1000,
                        background_color ='white',
                     max_words = 120).generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilInear')
plt.axis('off')
plt.show()
```

We can get a general sense of the job descriptions.

- Customer service looks like a job mostly posted. sales team also has openings.
- Communication skill, fast paced, bachelor degree are needed.
- Most companies are looking for diversity and equal opportunities.
